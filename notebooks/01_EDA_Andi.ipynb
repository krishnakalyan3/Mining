```python
def merge_dfs(left, right, left_on, right_on=None, suffix="_d"):
    if right_on is None: right_on = left_on
    return left.merge(right, how="left", left_on=left_on, right_on=right_on, suffixes=("", suffix))
```

```python
data_complete = merge_dfs(activities, cases, "Case-Id", "id")
```

```python
data_complete = merge_dfs(data_complete, crp, "Event-Id", "event_id")
```

```python
data_complete = merge_dfs(data_complete, diagnose, "Event-Id", "event_id")
```

```python
data_complete = merge_dfs(data_complete, diagnostics, "Event-Id", "event_id")
```

```python
data_complete = merge_dfs(data_complete, lacticacid, "Event-Id", "event_id")
```

```python
data_complete = merge_dfs(data_complete, leucocytes, "Event-Id", "event_id")
```

```python
data_case_id_reindex = data_complete.set_index(["Case-Id", "sorting"])
data_case_id_reindex.sort_index(inplace=True)
```

```python
data_case_id_reindex.head()
```

```python
def get_case_KPIs(dataframe):
    df = {}
    df["number_of_steps"] = len(dataframe)
    df["duration"] = dataframe.timestamp.iloc[-1] - dataframe.timestamp.iloc[0]
    return pd.Series(df)
```

```python
case_KPIs = data_case_id_reindex.groupby(["Case-Id"]).apply(get_case_KPIs)
```
