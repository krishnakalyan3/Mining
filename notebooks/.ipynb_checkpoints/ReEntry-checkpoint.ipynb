```python
all_columns = select_col1 + select_col2 + ['target']
```

```python
data['Event-Name'].unique()
```

```python
select_col_re_entey = ['Event-Name', 'Case-Id']
df2= data[select_col_re_entey]
```

```python
re_entry_case_id = df2[df2['Event-Name'] == 'Return ER']['Case-Id'].unique()
```

```python
def gen_target(data):
    if data in re_entry_case_id:
        return 1
    else:
        return 0
```

```python
df['target']=  df['Case-Id'].apply(lambda x : gen_target(x))
```
